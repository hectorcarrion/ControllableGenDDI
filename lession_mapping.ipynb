```python
import numpy as np
from skimage.measure import regionprops
from skimage.morphology import rectangle, erosion
import random

def randomly_shift_lesion_with_padding(
    lesion_mask,
    healthy_mask,
    pad=5,
    image_to_shift=None,
    seed=None,
    ensure_within_image=False
):
    """
    Randomly move a binary lesion_mask within the bounds of a healthy_mask,
    ensuring a 'pad' pixel border from the edge of the healthy region.

    Optionally, shift an accompanying 'image_to_shift' by the same offset
    (filling newly exposed pixels with 0).

    Parameters
    ----------
    lesion_mask : (H, W) ndarray (bool or 0/1)
        The binary mask for the lesion.
    healthy_mask : (H, W) ndarray (bool or 0/1)
        The binary mask for the healthy skin area.
    pad : int
        The desired minimum number of pixels between the lesion and the
        boundary of the healthy_mask.
    image_to_shift : (H, W) or (H, W, C) ndarray, optional
        An additional image to shift by the same offset. Pixels that move
        "out of frame" are discarded; newly exposed pixels are filled with 0.
        Must have the same height/width as lesion_mask.
    seed : int, optional
        If given, sets the random seed for reproducibility.

    Returns
    -------
    shifted_lesion_mask : (H, W) ndarray (bool)
        A new mask where the lesion has been randomly placed within the
        healthy region, with 'pad' pixels of padding from the boundary.
    shifted_image : ndarray, optional
        If `image_to_shift` is provided, returns the shifted image. Otherwise,
        returns None.
    """
    if image_to_shift:
        image_to_shift = np.array(image_to_shift)

    if seed is not None:
        random.seed(seed)

    # Ensure boolean
    lesion_mask = lesion_mask.astype(bool)
    healthy_mask = healthy_mask.astype(bool)

    # For convenience
    H, W = healthy_mask.shape

    # ----------------------------
    # 1) Find lesion bounding box
    # ----------------------------
    props = regionprops(lesion_mask.astype(int))
    if len(props) == 0:
        raise ValueError("No lesion found in the lesion_mask.")
    # If multiple lesions, we use the first region.
    minr, minc, maxr, maxc = props[0].bbox  # (min_row, min_col, max_row, max_col)
    lesion_h = maxr - minr
    lesion_w = maxc - minc

    # ---------------------------------
    # 2) Erode healthy_mask so lesion fits
    # ---------------------------------
    selem_fit = rectangle(lesion_h, lesion_w)
    fit_mask = erosion(healthy_mask, selem_fit)

    # -----------------------------------------
    # 3) Additional erosion for 'pad' distance
    # -----------------------------------------
    if pad > 0:
        selem_pad = rectangle(2 * pad, 2 * pad)
        padded_fit_mask = erosion(fit_mask, selem_pad)
    else:
        padded_fit_mask = fit_mask

    # All candidate top-left corners that keep the lesion inside the healthy region
    valid_coords = np.argwhere(padded_fit_mask == 1)

    # ----------------------------------------------------------------
    # 4) Filter out corners that would go outside the *image* boundary
    # ----------------------------------------------------------------
    if ensure_within_image:
        valid_coords_filtered = []
        for y, x in valid_coords:
            # Make sure the entire bounding box stays in [0..H) x [0..W)
            if (y + lesion_h <= H) and (x + lesion_w <= W):
                valid_coords_filtered.append((y, x))

        valid_coords_filtered = np.array(valid_coords_filtered)
        if len(valid_coords_filtered) == 0:
            raise ValueError(
                "No valid placement found that fits inside both the healthy region "
                "and the image boundaries (lesion might be too large or 'pad' too big)."
            )
    else:
        valid_coords_filtered = valid_coords

    # -----------------------
    # 5) Pick a random point
    # -----------------------
    rand_idx = random.randint(0, len(valid_coords_filtered) - 1)
    top_left_y, top_left_x = valid_coords_filtered[rand_idx]

    # -------------------------------------
    # 6) Create a new lesion mask, shifted
    # -------------------------------------
    shifted_lesion_mask = np.zeros_like(lesion_mask, dtype=bool)

    # Crop out the lesion region from the bounding box
    lesion_region = lesion_mask[minr:maxr, minc:maxc]

    # Place it in the chosen location
    shifted_lesion_mask[
        top_left_y : top_left_y + lesion_h,
        top_left_x : top_left_x + lesion_w
    ] = lesion_region

    # ------------------------------------------------
    # 7) (Optional) Shift an additional image similarly
    # ------------------------------------------------
    shifted_image = None
    if image_to_shift is not None:
        offset_y = top_left_y - minr
        offset_x = top_left_x - minc
        shifted_image = _shift_array(image_to_shift, offset_y, offset_x)
        shifted_image = Image.fromarray(shifted_image)

    # --------------------------------------------------------------
    # 8) (Optional) Validate lesion is indeed inside healthy region
    # --------------------------------------------------------------
    if ensure_within_image:
        if np.any(shifted_lesion_mask & ~healthy_mask):
            raise ValueError("Random shift placed lesion partially outside healthy area!")

    return shifted_lesion_mask, shifted_image

def _shift_array(arr, shift_y, shift_x):
    """
    Shift a 2D or 3D array (H,W) or (H,W,C) by (shift_y, shift_x) pixels.
    Pixels that move out of the frame are discarded.
    Newly exposed pixels are filled with zero.
    """
    out = np.zeros_like(arr)

    if arr.ndim == 2:
        H, W = arr.shape
        src_y_start, src_y_end, dst_y_start, dst_y_end = _compute_shift_bounds(H, shift_y)
        src_x_start, src_x_end, dst_x_start, dst_x_end = _compute_shift_bounds(W, shift_x)

        out[dst_y_start:dst_y_end, dst_x_start:dst_x_end] = arr[src_y_start:src_y_end, src_x_start:src_x_end]

    else:  # (H, W, C)
        H, W, C = arr.shape
        src_y_start, src_y_end, dst_y_start, dst_y_end = _compute_shift_bounds(H, shift_y)
        src_x_start, src_x_end, dst_x_start, dst_x_end = _compute_shift_bounds(W, shift_x)

        out[dst_y_start:dst_y_end, dst_x_start:dst_x_end, :] = arr[src_y_start:src_y_end, src_x_start:src_x_end, :]

    return out

def _compute_shift_bounds(size, shift):
    """
    Compute slicing bounds for a single dimension of length 'size',
    given an integer shift. If shift > 0, data is moved 'down/right'.
    """
    if shift >= 0:
        # Shifting "forward"
        src_start = 0
        src_end   = size - shift
        dst_start = shift
        dst_end   = size
    else:
        # Shifting "backward" (shift < 0)
        src_start = -shift
        src_end   = size
        dst_start = 0
        dst_end   = size + shift  # shift is negative => smaller
    return src_start, src_end, dst_start, dst_end
```

```python
for img in tqdm(image_files):
    init_image = load_image(img).convert('RGB') # healhty
    image_out = load_image(seg_base + img.split('/')[-1]) # diseased
    image_out_og = load_image(seg_base + img.split('/')[-1]) # diseased

    # get mask
    segmentation_map = np.load(seg_base + img.split('/')[-1].replace('.png', '.npy'))
    lesion_mask = ((segmentation_map == 1)).astype(np.float32)
    skin_mask = ((segmentation_map == 4)).astype(np.float32)

    # Add the lesion to the valid skin position
    skin_mask = skin_mask + lesion_mask

    # no blur no dialation yet
    kernel = np.ones((3, 3), np.uint8)
    lesion_mask = cv2.dilate(lesion_mask, kernel, iterations=2)

    # move around
    try:
        shifted_lesion_mask, image_out = randomly_shift_lesion_with_padding(lesion_mask, skin_mask,
                                                             pad=10, image_to_shift=image_out,
                                                             seed=None, ensure_within_image=True)
    except:
        try:
            shifted_lesion_mask, image_out = randomly_shift_lesion_with_padding(lesion_mask, skin_mask,
                                                                pad=5, image_to_shift=image_out,
                                                                seed=None, ensure_within_image=True)
        except:
            try:
                shifted_lesion_mask, image_out = randomly_shift_lesion_with_padding(lesion_mask, skin_mask,
                                                    pad=0, image_to_shift=image_out,
                                                    seed=None, ensure_within_image=True)
            except:
                print("No region found")
                shifted_lesion_mask = lesion_mask

    lesion_mask_pil = Image.fromarray((shifted_lesion_mask * 255).astype(np.uint8), mode="L")
    skin_mask_pil = Image.fromarray((skin_mask * 255).astype(np.uint8), mode="L")

    lesion_mask_pil = blur_pipeline.mask_processor.blur(lesion_mask_pil, blur_factor=2)

    overlayed_img = pipeline_sd.image_processor.apply_overlay(lesion_mask_pil,
                                                              init_image,
                                                              image_out)

    #overlayed_img = overlay_lesion_on_skin(init_image, image_out, lesion_mask, skin_mask, pad=10, seed=42)

    # Create a figure and axes manually
    fig, axes = plt.subplots(1, 5, figsize=(16, 4))

    axes[0].imshow(init_image)
    axes[0].set_title("Healhty Image")
    axes[0].axis("off")

    axes[1].imshow(image_out_og)
    axes[1].set_title("Original Image")
    axes[1].axis("off")

    axes[2].imshow(lesion_mask_pil)
    axes[2].set_title("Shifted Lesion Mask")
    axes[2].axis("off")

    axes[3].imshow(skin_mask_pil)
    axes[3].set_title("Healthy Skin Mask")
    axes[3].axis("off")

    axes[4].imshow(overlayed_img)
    axes[4].set_title("Generated Image")
    axes[4].axis("off")


    plt.savefig(f"/content/drive/MyDrive/ucsc_drive/UCSC/Winter 2023/DDI/Synthetics/Lesion Shift/Run 2/{img.split('/')[-1][:-4]}.pdf")
    plt.show()  # Ensure the plot is displayed
    plt.close(fig)  # Close figure to free memory

    overlayed_img.save(f"/content/drive/MyDrive/ucsc_drive/UCSC/Winter 2023/DDI/Synthetics/Lesion Shift/Run 2/{img.split('/')[-1]}")
```

```python
def find_centroid(binary_mask):
    indices = np.argwhere(binary_mask == 1)  # Get (row, col) positions of 1s
    if indices.size == 0:  # Handle empty case
        return None
    centroid = indices.mean(axis=0)  # Compute centroid
    centroid_int = np.round(centroid).astype(int)  # Round to nearest integer
    return tuple(centroid_int[::-1])  # Swap (row, col) -> (x, y)
```
